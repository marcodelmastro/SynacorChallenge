```python
def readInput(filename="synacor-challenge/challenge.bin"):
    prog = []
    i = 0
    with open(filename, "rb") as f:
        read = f.read(2)
        while read:
            prog.append( int.from_bytes(read, "little") )
            read = f.read(2)
    return prog
```

```python
mem = readInput()
print(len(mem))
```

```python
import pickle

with open("test.sav", "wb") as f:
    pickle.dump(mem,f)
```
